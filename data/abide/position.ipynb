```python
import os

def generate_file_list():
    # 1. 定义源目录路径
    source_dir = '/mnt/dataset4/DATASETS/fmri_pretraining/preprocess_z-score/WAR_NPYZ/ABIDE_val'
    
    # 2. 定义输出 txt 文件路径
    output_file = '/home/chenx/code/neurostorm_ncc/data/abide/abide_val.txt'

    # 3. 确保输出文件的父目录存在，如果不存在则创建
    output_dir_path = os.path.dirname(output_file)
    if not os.path.exists(output_dir_path):
        os.makedirs(output_dir_path)
        print(f"已创建目录: {output_dir_path}")

    # 用于存储所有找到的文件路径
    file_paths = []

    # 4. 遍历源目录
    # os.walk 可以递归查找子文件夹。如果你确定只有一层目录，也可以用这个，兼容性更好。
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            # 拼接成绝对路径
            full_path = os.path.join(root, file)
            file_paths.append(full_path)

    # 5. 排序 (非常重要：保证训练数据顺序固定)
    file_paths.sort()

    # 6. 写入文件
    try:
        with open(output_file, 'w') as f:
            for path in file_paths:
                f.write(path + '\n')
        
        print(f"成功！已将 {len(file_paths)} 个文件路径写入到:")
        print(output_file)
        
    except Exception as e:
        print(f"写入文件时发生错误: {e}")

if __name__ == "__main__":
    generate_file_list()
```
